```python
sns.boxplot(x='State', y='Electric Range', data=df).set(title='State vs Electric Range')
```

```python
import plotly.express as px

# Aggregating data by state
state_counts = df.groupby('State').size().reset_index(name='EV Count')

# Create choropleth map
fig = px.choropleth(state_counts,
                    locations='State',
                    color='EV Count',
                    hover_name = 'EV Count',
                    locationmode='USA-states',
                    animation_frame ='EV Count')
fig.show()
```

```python
df_racingBarPlot = df.groupby(["Model Year", "Make"]).size().reset_index(name="EV Count")
df_racingBarPlot_pivot = df_racingBarPlot.pivot(index="Model Year", columns="Make", values="EV Count").fillna(0)
df_racingBarPlot.head()
```

```python
# Create the racing bar plot without fig_kwargs in bar_chart_race
bcr.bar_chart_race(
    df=df_racingBarPlot_pivot,
    filename="final.mp4",
    fixed_max = True,
    n_bars=10,
    filter_column_colors = True,
    period_length=3000,
    title='Year wise sales of each Makers from last decade')
```
