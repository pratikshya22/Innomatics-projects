### About the Dataset
##### The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate.

# Data Preprocessing

```python
# Convert date columns to datetime format
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
df['DOJ'] = pd.to_datetime(df['DOJ'], errors='coerce')
df['DOL'] = pd.to_datetime(df['DOL'], errors='coerce')
df.info()
```

```python
#Finding total zero and -1 in data
zero_one = (df==0).sum() + (df==-1).sum()
print(zero_one)
```

```python
# finding percentage of zero and -1 in data
zero_one_percent = (zero_one/df.shape[0])*100
print(zero_one_percent)
```

```python
zero_one = (df==0).sum() + (df==-1).sum()
print(zero_one)
```

```python
# deleting the columns with more than 80% of zero and -1
df.drop(columns=["MechanicalEngg", "ElectricalEngg", "TelecomEngg", "CivilEngg"], inplace=True)
df.head()
```

```python
# Converting 0 and -1 of the data to null as per description except "CollegeCityTier"
cols_to_replace = df.columns.difference(['CollegeCityTier'])
df[cols_to_replace] = df[cols_to_replace].replace({0: np.nan, -1: np.nan})
df.head()
```

```python
#"Date of Joining" Changing Datatype from Object to DateTime
df['DOJ'] = pd.to_datetime(df['DOJ'])
```

##### Is there a relationship between gender and specialization?


*   Using Chi-Square Test of Independence to check if gender influences specialization choices.



```python
from scipy.stats import chi2_contingency

# Creating a contingency table for Gender and Specialization
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Applying the Chi-Square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Test p-value: {p_value}")
```

Observation: The Chi-Square test reveals a significant relationship between gender and specialization, as the p-value is far below 0.05. This indicates that specialization preferences vary significantly between male and female students.
